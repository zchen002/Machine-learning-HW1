# Logistic regression model

```
def sigmoid(z):

    s = 1/(1 + np.exp(-z))
    
    return s
```

```
def initialize_params(dim):

    # dim -- number of parameters
    
    w = np.zeros((dim, 1))
    b = 0.0

    assert(w.shape == (dim, 1))
    assert(isinstance(b, float) or isinstance(b, int))
    
    return w, b
```

**Weighted binary cross entropy loss for imbalanced training data**

```
def propagate(w, b, X, Y):
    
    # cost -- binary cross entropy loss for logistic regression
    # dw -- gradient of the loss with respect to w, thus same shape as w
    # db -- gradient of the loss with respect to b, thus same shape as b
    
    
    m = X.shape[1]

    # 1. FORWARD PROPAGATION

    A = sigmoid(np.dot(w.T, X) + b) # compute activation
    #print(A.shape)
    
    # weighted cross entropy = -w0 * ylog(a) - w1 * (1-y)log(1-a), where w0 and w1 are the weights for class 1 and 0, respectively
    # in this case w0 = 9, w1 = 1
    cost = -1/m * (9 * np.dot(Y, np.log((A + 1e-15).T)) + 1 * np.dot(1 - Y, np.log(((1 - A) + 1e-15).T))) # numerical safe     

    # 2. BACKWARD PROPAGATION
    # Compute gradients of loss function
    dw = 1/m * np.dot(X, (A + 8*Y*A - 9*Y).T)
    db = 1/m * np.sum(A + 8*Y*A - 9*Y)

    #print(dw.shape)
    #print(w.shape) # dw.shape = w.shape 
    assert(dw.shape == w.shape) #telling the program to test that condition, and trigger an error if the condition is false
    assert(db.dtype == float)
    cost = np.squeeze(cost)
    assert(cost.shape == ())
    grads = {"dw": dw,
             "db": db}
    
    return grads, cost
```

**Mini-batch gradient descent**

```
def optimize(w, b, X, Y, num_iters, learning_rate, batch_size, print_cost=False):
    
    # X, Y: training data and training labels
    
    costs = []
    
    for iter in range(num_iters):
        # Randomize training data
        X, Y = shuffle(X.T, Y.T)
        minibatch_size = batch_size
        #if iter == 500:
          #print(X.shape)
          #print(Y.shape)
          
        # mini batch  
        for i in range(0, X.shape[0], minibatch_size):   
            X_mini = X[i:i + minibatch_size]
            Y_mini = Y[i:i + minibatch_size]
        #if iter == 500:
          #print(X_mini.shape)
          #print(Y_mini.shape)
 
        grads, cost = propagate(w, b, X_mini.T, Y_mini.T)                             
        
       
        # Retrieve gradients
        dw = grads["dw"]
        db = grads["db"]
        
        # Update w, b using gradient descent 
        w = w - learning_rate * dw                    
        b = b - learning_rate * db
        
        X, Y = X.T, Y.T
        #if iter == 500:
          #print(X.shape)
          #print(Y.shape)
          
        # Record the costs
        if iter % 100 == 0:
            costs.append(cost)
        
        # Print the cost every 200 training examples
        if print_cost and (iter % 200 == 0 or iter == num_iters - 1):
            print ("Cost after iteration %i: %f" % (iter, cost))
    
    params = {"w": w,
              "b": b}
    
    grads = {"dw": dw,
             "db": db}
    
    return params, grads, costs
```

```
def predict(w, b, X):
    
    m = X.shape[1]
    Y_pred = np.zeros((1, m)) # to store predictions
    
    # Apply w and b that are learned through training
    A = sigmoid(np.dot(w.T, X) + b) 

    # A[0][i] = A[0,i]
    for i in range(A.shape[1]):
        # Convert probabilities a[0,i] to actual predictions p[0,i] (0 or 1)
        if(A[0,i] > 0.5):
           Y_pred[0,i] = 1 
        else:    
           Y_pred[0,i] = 0 
            
    
    assert(Y_pred.shape == (1, m))
    
    return Y_pred
```

```
def model(X_train, Y_train, X_test, Y_test, num_iters=2000, learning_rate=0.005, batch_size=32, print_cost=False):
        
    #Initialize parameters (with zeros)
    w, b = initialize_params(X_train.shape[0])                            

    # Gradient descent
    parameters, grads, costs = optimize(w, b, X_train, Y_train, num_iters, learning_rate, batch_size, print_cost)   
    
    # Retrieve trained w and b
    w = parameters["w"]
    b = parameters["b"]
    
    # Predict on test/train set
    Y_pred_test = predict(w, b, X_test)
    Y_pred_train = predict(w, b, X_train)


    # Print train/test accuracy
    print("")
    print("train accuracy: {} %".format(100 - np.mean(np.abs(Y_pred_train - Y_train)) * 100))
    print("test accuracy: {} %".format(100 - np.mean(np.abs(Y_pred_test - Y_test)) * 100))
    #print(w)
    #print(b)

    # d -- dictionary storing information about the model
    d = {"costs": costs,
         "Y_pred_test": Y_pred_test, 
         "Y_pred_train" : Y_pred_train, 
         "w" : w, 
         "b" : b,
         "learning_rate" : learning_rate,
         "num_iters": num_iters}
    
    return d
```

# Applying classifiers 0-9 together on test set (argmax)

```
prediction = []

for col_id in range(test_images.shape[1]):
  column = test_images[:, col_id]
  #print(column.shape)
  column = column.reshape((column.shape[0], 1))
  #print(column.shape)
  tmp = []
  # Apply classifiers 0-9
  tmp.append(sigmoid(np.dot(w0.T, column) + b0)[0,0])
  tmp.append(sigmoid(np.dot(w1.T, column) + b1)[0,0])
  tmp.append(sigmoid(np.dot(w2.T, column) + b2)[0,0])
  tmp.append(sigmoid(np.dot(w3.T, column) + b3)[0,0])
  tmp.append(sigmoid(np.dot(w4.T, column) + b4)[0,0])
  tmp.append(sigmoid(np.dot(w5.T, column) + b5)[0,0])
  tmp.append(sigmoid(np.dot(w6.T, column) + b6)[0,0])
  tmp.append(sigmoid(np.dot(w7.T, column) + b7)[0,0])
  tmp.append(sigmoid(np.dot(w8.T, column) + b8)[0,0])
  tmp.append(sigmoid(np.dot(w9.T, column) + b9)[0,0])
  #print(tmp)
  # Determine the classifier with the highest probability and declare the corresponding digit as prediction
  prediction.append(np.argmax(tmp))
  
#print(prediction)
prediction = np.asarray(prediction)
#print(prediction)
print(prediction.shape)
prediction = prediction.reshape((prediction.shape[0], 1))
print(prediction.shape)
print(test_labels.shape)

print("test accuracy: {} %".format(100 - np.mean(np.abs(prediction.T - test_labels)) * 100))
```
